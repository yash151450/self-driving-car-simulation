```python
import math
def grayscale(img):
    return cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

def canny(img,low_t,high_t):
    return cv2.Canny(img,low_t,high_t)

def blur(img,ksize):
    return cv2.GaussianBlur(img,(ksize,ksize),0)

def polygon(mask,pts):
    return cv2.fillPoly(mask,pts,(255,255,255))

def masking(img,masked):
    return cv2.bitwise_and(img,masked)
    
def hough_lines(img,rho,theta,threshold,min_line_length,max_line_gap):
    return cv2.HoughLinesP(img,rho,theta,threshold,np.array([]),
                          min_line_length,max_line_gap)
```

```python
def lanefinder(img):
    gray=grayscale(img)
    blur1=blur(gray,9)
    canny1=canny(blur1,100,200)
    pts=np.array([[[0,400],[0,250],[230,170],[430,170],[600,200],[600,400]]],dtype=np.int32)
    black=np.zeros(canny1.shape,dtype=np.uint8)
    masked=polygon(black,pts)
    f_masked=masking(canny1,masked)
    
    rho=2
    theta=np.pi/180
    threshold=20
    min_line_length=40
    max_line_gap=20
    lines=hough_lines(f_masked,rho,theta,threshold,min_line_length,max_line_gap)
    
    line_image = np.copy(image1)*0
    cv2.imshow('gray',gray)
    cv2.imshow('blur1',blur1)
    cv2.imshow('canny1',canny1)
    cv2.imshow('masked',masked)
    cv2.imshow('final_masked_image',f_masked)
    cv2.imshow('image 1',image1)
    #cv2.imshow('black',black)

    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(line_image,(x1,y1),(x2,y2),(0,0,255),7)
    
    color_edges=np.dstack((canny1,canny1,canny1))
    
    cv2.imshow('line_image',line_image)
    lines_edges=cv2.addWeighted(color_edges, 0.8, line_image, 1, 0)
    
    final_image=cv2.addWeighted(lines_edges,1,image1,1,0)
    return final_image
    #return gray,blur1,canny1,masked,f_masked,lines
```
